```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np 
import csv
df = pd.read_csv("hepsiemlakpreprocessing.csv", encoding="utf-8")
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='le', y='Fiyat')
plt.title('Konuma Gre Fiyat Dalm')
plt.xticks(rotation=90)
plt.show()
```

```python
avg_price_by_province = df.groupby('le')['Fiyat'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
avg_price_by_province.plot(kind='bar')
plt.title('Konumlara Gre Ortalama Fiyatlar')
plt.ylabel('Ortalama Fiyat')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter


df = pd.read_csv("hepsiemlakpreprocessing.csv")  


plt.figure(figsize=(10, 6))
sns.boxplot(x='Oda Says', y='Fiyat', data=df)


plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f'{x:,.0f}'))

plt.title('Fiyat Aralna Gre Oda Says Dalm')
plt.xlabel('Oda Says')
plt.ylabel('Fiyat (TL)')
plt.xticks(rotation=45)
plt.show()
```

```python

df['Net m2 Bin'] = pd.cut(df['Net m2'], bins=10)


plt.figure(figsize=(16, 10))
sns.boxplot(x='Net m2 Bin', y='Fiyat', data=df, palette='coolwarm')


plt.title('Fiyat Dalm iin Metrekare Aralklar', fontsize=16)
plt.xlabel('Metrekare Aral', fontsize=14)
plt.ylabel('Fiyat (TL)', fontsize=14)


plt.xticks(rotation=45)

plt.show()
```
